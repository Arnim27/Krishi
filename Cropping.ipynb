```python
import os
from PIL import Image

def resize_images(src_directory, dest_directory, size=(128,128)):
    os.makedirs(dest_directory, exist_ok=True)
    
    for filename in os.listdir(src_directory):
        if filename.endswith((".png", ".jpg", ".jpeg")):
            src_path = os.path.join(src_directory, filename)
            dest_path = os.path.join(dest_directory, filename)
            with Image.open(src_path) as img:
                img_resized = img.resize(size, Image.LANCZOS)
                img_resized.save(dest_path)
                print(f"Resized and saved: {dest_path}")

src_base = "Brain_Cancer"
dest_base = "cropped"

subdirectories = [
    "brain_glioma",
    "brain_menin",
    "brain_tumor"
]

for subdir in subdirectories:
    src_dir = os.path.join(src_base, subdir)
    dest_dir = os.path.join(dest_base, subdir)
    print(f"Processing directory: {src_dir}")
    resize_images(src_dir, dest_dir)
    print(f"Finished processing {src_dir}")

print("All images have been resized to 128*128 and saved in the 'cropped' directory.")
```
